We have covered various models for classification and regression which allow us to fit a response variable to a set of feature variables. Usually, all models have a variety of parameters which can be tuned to find the best among all possible models. Best in that context means not only fitting the training data well, but generalising also to previously unseen test data. 

In the lessons you have seen how to use regularisation and grid search in combination with cross validation to find a good model according to model scores like MSE and R2 in the case of regression, and accuracy, recall or precision in the case of classification.

As a reference below we will illustrate this procedure for a binary classification problem with logistic regression.

The same steps can be used for other models if you replace with the appropriate tuning parameters.
 

We assume you already have a data set with the features and the target you want to model (this is the output you get from doing the EDA).

Then the basic workflow generalises in the following way:
-----

1. Create an Instance of the model

1. Check for the parameters of the model (each model will come with its own particular set of parameters)

1. Set up a search grid for the parameters you want to tune in form of a dictionary with the tuning parameters as keys and lists of possible values

1. Call GridSearchCV with the model, the parameter grid, the number of cross validation steps and the scoring of your choice

1. Fit the model

1. Extract the best score

1. Extract the best parameters

1. Choose the model with the best parameters and predict target variable values (here chosen for some of the values in the dataset)

## Example

First we load the libraries that we will need. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.5)
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
from sklearn import datasets, metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

## Get the data

### Load the dataset included in sklearn

- Choose breast cancer for binary classification
- Choose iris for multiclass classification

```python
# Choose one of the two datasets

data = datasets.load_breast_cancer()
#data = datasets.load_iris()
```

### Load the feature matrix

```python
X = pd.DataFrame(data.data, columns=data.feature_names)

X.head(3)
```

### Load the target variable 

```python
y = pd.Series(data.target)
y.head(3)
```

### Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                        stratify=y, test_size=0.3, random_state=1)
```

### Rescale the features

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

## Logistic regression

Now that we have the data (features and target), we will go through the steps in the workflow, so we can fit a linear regression on _**all**_ of our features.

### Create an instance of the model (here you could also use other classifiers)

```python
model = LogisticRegression(solver='liblinear', multi_class='ovr')
```

### Check for the parameters of the model (each model will come with its own particular set of parameters)

```python
list(model.get_params().keys())
```

**Note:** Not all of the model parameters are tuning parameters. 'C', 'penalty', 'fit_intercept' for example are tuning parameters, whereas 'n_jobs', 'verbose' or 'random_state' are not. 

### Set up a search grid for the parameters you want to tune in form of a dictionary with the tuning parameters as keys and lists of possible values

```python
params = {'C': np.logspace(-5, 5, 15),
          'penalty': ['l1', 'l2'],
          'fit_intercept': [True, False]}
```

### Call GridSearchCV with the model, the parameter grid, the number of cross validation steps and the scoring of your choice

Setting `return_train_score` to `True` will store more results.

```python
gs = GridSearchCV(estimator=model,
                  param_grid=params,
                  cv=5,
                  scoring='accuracy',
                  return_train_score=True)
```

### Fit the model

```python
gs.fit(X_train, y_train)
```

### Extract the best model

```python
gs_best = gs.best_estimator_
gs_best
```

### Extract the scores

#### Cross validated training score of the best model

This score results as the best average of cross validation scores for any of the particular model settings.

```python
print(gs.best_score_)
```

This is equivalent to:

```python
cross_val_score(gs.best_estimator_, X_train, y_train, cv=5).mean()
```

#### Score on the whole training set

This is the score resulting with the best model fit on the whole training set (without cross validation).

```python
gs.score(X_train, y_train)
```

#### The score of the best model on the test set

```python
gs.score(X_test, y_test)
```

### Extract the best parameters

```python
print(gs.best_params_)
```

### Choose the model with the best parameters and predict target variable values (here chosen for some of the values in the dataset)

```python
gs.predict(X_test.iloc[10:20])
```

### Further analysis

#### Get all results from the grid search (they are returned as a dictionary that you can conveniently read into a data frame)

```python
results = pd.DataFrame(gs.cv_results_)
results
```

#### The most interesting result is the mean test score for each parameter combination

```python
results[['mean_test_score'] +
        [col for col in results.columns if 'param_' in col]].head()
```

```python
grouped = results.groupby(['param_penalty', 'param_fit_intercept'])
groups = list(grouped.groups.keys())

for i in range(len(groups)):
    grouped.get_group(groups[i]).plot(
        x='param_C', y='mean_test_score', logx=True, lw=3, c='b', title='Penalty: {} , Intercept: {}'.format(groups[i][0], groups[i][1]))
plt.show()
```

## Code summary 

```python
# prepare the data

X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)

X_train, X_test, y_train, y_test = train_test_split(X, y,
                        stratify=y, test_size=0.3, random_state=1)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
# choose and inspect the model

model = LogisticRegression(solver='liblinear', multi_class='ovr')
list(model.get_params().keys())
```

```python
# setup the grid search

params = {'C': np.logspace(-4, 4, 10),
          'penalty': ['l1', 'l2'],
          'fit_intercept': [True, False]}

gs = GridSearchCV(estimator=model,
                  param_grid=params,
                  cv=5,
                  scoring='accuracy',
                  return_train_score=True)

gs.fit(X_train, y_train)

# extract the grid search results

print('Best Parameters:')
print(gs.best_params_)
print('Best estimator C:')
print(gs.best_estimator_.C)
print('Best estimator mean cross validated training score:')
print(gs.best_score_)
print('Best estimator score on the full training set:')
print(gs.score(X_train, y_train))
print('Best estimator score on the test set:')
print(gs.score(X_test, y_test))
print('Best estimator coefficients:')
print(gs.best_estimator_.coef_)
```

> **Note:** This grid searched model will only select a single `C` even in the multiclass classification problem, i.e. the same `C` for all one-versus-rest models.

## Use `LogisticRegressionCV` in combination with gridsearch

```python
# choose and inspect the model

model = LogisticRegressionCV(
    solver='liblinear', multi_class='ovr', cv=5, max_iter=1000)
list(model.get_params().keys())
```

```python
# setup the grid search

params = {'penalty': ['l1', 'l2'],
          'Cs': [np.logspace(-4, 4, 10)],
          'fit_intercept': [True, False],
          'solver': ['liblinear']}

gs = GridSearchCV(estimator=model,
                  param_grid=params,
                  cv=5,
                  scoring='accuracy',
                  return_train_score=True)

gs.fit(X_train, y_train)

# extract the grid search results

print('Best Parameters:')
print(gs.best_params_)
print('Best estimator C:')
print(gs.best_estimator_.C_)
print('Best estimator mean cross validated training score:')
print(gs.best_score_)
print('Best estimator score on the full training set:')
print(gs.score(X_train, y_train))
print('Best estimator score on the test set:')
print(gs.score(X_test, y_test))
print('Best estimator coefficients:')
print(gs.best_estimator_.coef_)
```

> **Note:** This grid searched model will select an optimal `C` for each one-versus-rest model.

## Model evaluation

```python
import scikitplot as skplt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(sns.color_palette("husl", len(gs.classes_)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

### Confusion matrix

```python
predictions_train = gs.predict(X_train)
predictions_test = gs.predict(X_test)
```

```python
skplt.metrics.plot_confusion_matrix(y_train, predictions_train, figsize=(6,6))
plt.ylim([-0.5, len(y_train.unique())-0.5])
plt.show()
# Note -  you can normalize this with normalize=True
```

```python
skplt.metrics.plot_confusion_matrix(y_test, predictions_test)
plt.ylim([-0.5, len(y_test.unique())-0.5])
plt.show()
```

### Classification report

```python
print(classification_report(y_train, predictions_train))
```

```python
print(classification_report(y_test, predictions_test))
```

### Precision-recall and ROC curve

```python
probabilities_train = gs.predict_proba(X_train)
probabilities_test = gs.predict_proba(X_test)
```

```python
skplt.metrics.plot_precision_recall(y_train, probabilities_train, cmap=cmap)
plt.show()
```

```python
skplt.metrics.plot_roc(y_train, probabilities_train, cmap=cmap)
plt.show()
```
