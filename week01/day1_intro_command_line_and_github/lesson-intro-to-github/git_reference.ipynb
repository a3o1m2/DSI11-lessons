# Git reference
---

### Clone a repository:

```bash
git clone -o ref_name url_of_repository
```

- `url_of_repository`: e.g. https://git.generalassemb.ly/DSI-LDN-11/DSI11-lessons.git 
- `ref_name`: name with which you are going to refer to a repository, default: `origin`
    - common `ref_name` conventions: 
        - `upstream` for https://git.generalassemb.ly/DSI-LDN-11/DSI11-lessons.git 
        - `origin` for your forked version of the repository at https://git.generalassemb.ly/[your_github_handle]/DSI11-lessons 
        
- Example:
    ```bash 
git clone -o upstream https://git.generalassemb.ly/DSI-LDN-11/DSI11-lessons 
```
 
**Important**: Never clone from a repository which is already tracked by git. Nested repositories will create problems!!!
               
### Add other forks as remotes referring to them with `ref_name`:

```bash
git remote add ref_name url
```     
        
### Check which remote repositories are set:

```bash
git remote -v
```

- Example output:

```bash
origin	https://git.generalassemb.ly/[your_github_handle]/DSI11-lessons.git  (fetch)
origin	https://git.generalassemb.ly/[your_github_handle]/DSI11-lessons.git (push)
upstream	https://git.generalassemb.ly/DSI-LDN-11/DSI11-lessons.git (fetch)
upstream	https://git.generalassemb.ly/DSI-LDN-11/DSI11-lessons.git (push)
```

### Change the url for a `ref_name` to a particular repository:

```bash
git remote set-url ref_name new_url
```

### Remove the reference to a repository:

```bash
git remote rm ref_name
```

### Pull from a repository (you need read permissions):

```bash
git pull ref_name master
```

- `master` refers to the default git branch where you are writing to (we will learn more about branches at a later stage)

### Push to a repository (you need write permissions):

```bash
git push ref_name master
```

### Add files in the current directory to the staging area:

```bash
git add .
```

### Add a particular file (or a folder) to the staging area:

```bash
git add file_name
```

### Commit changes:

```bash
git commit -m 'your commit message'
```


### General rule:

**Add your file changes frequently to the staging area, commit the changes (with a meaningful but short message) and push to your personal repository - it is like a backup.**
