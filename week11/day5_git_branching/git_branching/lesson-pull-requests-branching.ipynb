## Pull request 

### What is a pull request ?

- After adding, committing, and pushing, if you're collaborating on a project, one of the common ways to **share your code** is to do a pull request (PR) 
- At that point, whoever you're working with **will look over the code you've issued a pull request on**, and make any comments/suggestions they have, or ask any questions
    - If hosted through Github (or some similar system), these typically **take place through the browser**. In it you can have a discussion about pieces of the code and even view the changes made (see the pictures in the following slides).
- Once your code is ready to merge, the other person (or you if you have the correct rights) will merge via the browser (ie: "accept" the Pull Request, and then close it).

### Let's practice branching

#### Create a repository

- Go to https://git.generalassemb.ly 
- Create a new repository (Click on the + in top right corner) and give it a name (note that you can make it public or private)
- Optional: Initialize the repository with a readme-file
- Optional: Add a `.gitignore` file designed for python
    - A `.gitignore` file can list files which should not be tracked (e.g. a lot of hidden files)
- Click on `clone or download` and copy the link

- Open the terminal and type `cd` + enter (making sure you are in your home directory)
- Enter: `git clone` + the link you just copied
- Type `ls` to see your repository and `cd` into it

HELP on e.g. commits = git help commit

Note - once you cd to the newly created folder...
- 'git remote -v' to see where you cloned it from.
- ls -a to see files inc hidden files
- the .git folder holds the updates information, it ensures tracking.
- you can vim the README to make updates
- then git status to see it's been modified
- then git add README.md, then git commit with a message

- If you have the local version, and the remote version, and you make different changes to it that conflict. Then 
when you come to pull the remote version down to your PC you will get a conflict you need to resolve manually
through vim and editing the README to the correct final version. Then you can git add, git commit, git push to unify changes.

- If you have made an update in remote repo but you'd like to check it out before merging etc... then you can do git checkout origin/master README.md (if you care about README file in this e.g.)

#### Create a branch

- Let's create a branch, type:  
    `git branch <name_of_your_branch>`
- We are now going to move the workspace to the branch we just created:  
    `git checkout <name_of_your_branch>`  
    
This is a different "working environment" than the master branch. As a consequence, if we **create files / folders and move back to the master branch, we won't be able to see them**. 

- Still in your branch, create a folder and a text file in it 
- Add, commit and push the folder you just created
- Note that you now need to specify the branch you are pushing to
- Next, go to your repository on https://git.generalassemb.ly and try to visualize the branch just created.

Finally let's merge your branch:
```
git checkout master
git pull origin master
git merge <name_of_your_branch>
git push origin master
```

Note - git diff master branch_1 gives the diff between two local branches

git checkout branch_1 enables you to change branch you're in

git branch gives list of branches and which one you're in|

git checkout branch_1 README.md will enable you to edit / checkout a file in another branch without having to switch to it.

you can also git push origin branch_1 which pushes that branch_1 onto github and you have the multiple branches

### The ugly duckling: merge conflict


Sometimes you get merge conflicts when merging or pulling from a branch. Git will then tell you something like

```
CONFLICT (content): Merge conflict in Fakefile
```

It also tells you to fix the conflicts and then to commit the result. **So how to fix the conflicts?**

### What is a merge conflict and how does it happen? 

A merge conflict usually occurs when your current branch and the branch you want to merge into the current branch **have diverged**. That is, you have commits in your current branch which **are not in the other branch, and vice versa**.

When Git merges the other branch into your current branch, it looks at the differences between latest common commits. When there are unambiguous differences (i.e. only one side changed a certain piece of code), **the changes are applied**.

The merge conflicts occur when there are **disagreeing changes** (typically, both branches have new commits). In that case, your conflicted file will have so-called conflict markers:

```
<<<<<<< HEAD
jars/VIB-lib.jar
=======
plugins/Plugin.jar
>>>>>>> branch-with-amazing-new-feature
```

### Resolving merge conflicts

#### First solution: tell git which file you want to keep

- for your version:
```
git checkout --ours <file_with_merge_conflict>
```

- for their version:
```
git checkout --theirs <file_with_merge_conflict>
```

#### Second solution: manually edit the file 

When you have a merge conflict, git will automatically point out the differences in the files using `<<<<<<` or `>>>>>>` (also called conflict markers) to indicate which changes belong to who.

- You can consequently either:
    - change manually the contents of the file 
    - use a built in tool like the one in `Atom`:
    


To see those, you just have to navigate to the file with merge conflicts and click on the changes you want to keep.
